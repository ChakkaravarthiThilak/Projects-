```python
import pandas as pd
```

```python
df=pd.read_csv(r'C:\Users\ELCOT\orders.csv')
df
```

```python
df.info()
```

```python
df.columns=df.columns.str.lower()
df.columns
```

```python
df.columns = df.columns.str.replace(' ', '_')
df
```

```python
type(df)
```

```python
df['sales_price'] = df['list_price'] * (1 - df['discount_percent'] / 100)
df['profit'] = df['sales_price'] - df['cost_price']
df['amount']=df['sales_price'] * df['quantity']
df
```

```python
df.rename(columns={'Old_Column_Name': 'New_Column_Name'}, inplace=True)
df
```

```python
df.to_csv(r'C:\Users\ELCOT\modified_orders.csv', index=False)
```
